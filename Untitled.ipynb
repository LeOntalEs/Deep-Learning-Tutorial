```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import Normalizer
```

```python
x = [k for k in range(0, 100)]
x = np.array(x)
```

```python
x = x.reshape([-1, 1])
print(x.shape)
```

```python
w, c = 2, 2

def f(x):
    return w*x*c

y = [f(k) for k in range(0, 100)]
y = np.array(y)
```

```python
print(y)
```

```python
plt.scatter(x, y)
```

```python
xt = [k for k in range(50, 150)]
xt = np.array(xt)
xt = xt.reshape([-1, 1])
yt = [f(k) for k in range(50, 150)]
yt = np.array(yt)
```

```python
est = LinearRegression().fit(x, y)
est.score(xt, yt)
```

```python
def f2(x):
    return np.sin(x)

x = [k for k in range(50, 150)]
x = np.array(x)
x = x.reshape([-1, 1])
y = [f2(k) for k in range(50, 150)]
y = np.array(y)

xt = [k for k in range(50, 150)]
xt = np.array(xt)
xt = xt.reshape([-1, 1])
yt = [f2(k) for k in range(50, 150)]
yt = np.array(yt)
```

```python
plt.plot(xt, yt)
```

```python
est = LinearRegression().fit(x, y)
est.score(xt, yt)
```
